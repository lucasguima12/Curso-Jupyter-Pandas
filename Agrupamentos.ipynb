## Criando Agrupamentos

```python
import pandas as pd
```

```python
dados = pd.read_csv('dados/aluguel_residencial.csv', sep = ';')
```

```python
dados
```

```python
dados['Valor'].mean()
```

```python
bairros = ['Barra da Tijuca ', 'Copacabana', 'Ipanema', 'Lebron', 'Botafogo', 'Flamengo', 'Tijuca']
```

```python
selecao = dados['Bairro'].isin(bairros)
dados = dados[selecao]
```

```python
grupo_bairro = dados.groupby('Bairro')
```

```python
grupo_bairro.groups
```

```python
for bairro, data in grupo_bairro:
    print(f'{bairro} tem valor medio de {data["Valor"].mean()}')
```

```python
grupo_bairro['Valor'].mean().round(2)
```

```python
alunos = pd.DataFrame({'Nome': ['Ary', 'Cátia', 'Denis', 'Beto', 'Bruna', 'Dara', 'Carlos', 'Alice'], 
                        'Sexo': ['M', 'F', 'M', 'M', 'F', 'F', 'M', 'F'], 
                        'Idade': [15, 27, 56, 32, 42, 21, 19, 35], 
                        'Notas': [7.5, 2.5, 5.0, 10, 8.2, 7, 6, 5.6], 
                        'Aprovado': [True, False, False, True, True, True, False, False]}, 
                        columns = ['Nome', 'Idade', 'Sexo', 'Notas', 'Aprovado'])
```

```python
alunos
```

```python
grupo = alunos.groupby('Sexo')
```

```python
grupo.groups
```

```python
grupo = alunos.groupby('Sexo')
grupo = pd.DataFrame(grupo['Notas'].mean().round(2))
grupo.columns = ['Notas Médias']
grupo
```
