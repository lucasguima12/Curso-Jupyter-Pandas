## Plot baixo nivel

```python
contagem_de_linguas = tmdb['original_language'].value_counts()
contagem_de_linguas = contagem_de_linguas.to_frame().reset_index().rename(columns={'index' : 'original_language', 'original_language' : 'total'})
contagem_de_linguas
```

```python
import seaborn as sns
```

```python
sns.barplot(x = 'original_language', y = 'total', data = contagem_de_linguas)
```

## Plot alto nivel

```python
sns.catplot(x = 'original_language', kind='count', data = tmdb)    ## o sns.catplot plota um grafico do df original, onde pega a coluna original language (eixo x) e conta quantas vezes cada valor aparece (com o kind = 'count')
```

```python
filmes_nao_ingles = tmdb.query('original_language != "en"')
total_filmes_nao_ingles = filmes_nao_ingles['original_language'].value_counts()
sns.catplot(x = 'original_language', 
            kind='count', 
            data = filmes_nao_ingles, aspect = 2, 
            order = total_filmes_nao_ingles.index, 
            palette = 'GnBu_d')
```
